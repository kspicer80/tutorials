```python
import pandas as pd
```

```python
dataset = pd.read_csv(r'.\datasets\chapter_3\nba_scores.csv', parse_dates=["Date"])
```

```python
dataset = dataset[['Date', 'Visitor/Neutral', 'VisitorPTS', 'Home/Neutral', 'HomePTS']]
dataset.head()
```

```python
dataset["HomeWin"] = dataset['VisitorPTS'] < dataset['HomePTS'] 
```

```python
y_true = dataset["HomeWin"].values
dataset['HomeLastWin'] = 0
dataset['VisitorLastWin'] = 0
```

```python
from collections import defaultdict
won_last = defaultdict(int)
```

```python
for index, row in dataset.iterrows():
    # if the dataframe is not sorted by 'Date,' instead pass dataset.sort('Date').iterrows()
    home_team = row['Home/Neutral']
    visitor_team = row['Visitor/Neutral']
    row['HomeLastWin'] = won_last[home_team]
    row['VisitorLastWin'] = won_last[visitor_team]
    dataset.loc[index] = row
    won_last[home_team] = row['HomeWin']
    won_last[visitor_team] = not row['HomeWin']
```

```python
won_last[home_team] = row['HomeWin']
won_last[visitor_team] = not row['HomeWin']
```

```python
dataset.loc[20:25]
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(random_state=14)
```

```python
X_previouswins = dataset[['HomeLastWin', 'VisitorLastWin']].values
```

```python
import numpy as np
from sklearn.model_selection import cross_val_score
scores = cross_val_score(clf, X_previouswins, y_true, scoring='accuracy')
print(f"The accuracy is {np.mean(scores) * 100} ")
```

```python
standings_filename = r'.\datasets\chapter_3\2021_2022_nba_standings.csv'
standings = pd.read_csv(standings_filename, skiprows=[0])
```

```python
standings
```

```python
dataset['HomeTeamRanksHigher'] = 0

for index, row in dataset.iterrows():
    home_team = row['Home/Neutral']
    visitor_team = row['Visitor/Neutral']
    home_rank = standings[standings['Team'] == home_team]['Rk'].values[0]
    visitor_rank = standings[standings['Team'] == visitor_team]['Rk'].values[0]
    row['HomeTeamRanksHigher'] = int(home_rank > visitor_rank)
    dataset.loc[index] = row
```

```python
X_homehigher = dataset[['HomeLastWin', 'VisitorLastWin', 'HomeTeamRanksHigher']].values
```

```python
clf = DecisionTreeClassifier(random_state=14)
scores = cross_val_score(clf, X_homehigher, y_true, scoring='accuracy')
print(f"The accuracy is {np.mean(scores) * 100} ")
```

```python
last_match_winner = defaultdict(int)
dataset['HomeTeamWonLast'] = 0

for index, row in dataset.iterrows():
    home_team = row['Home/Neutral']
    visitor_team = row['Visitor/Neutral']
    teams = tuple(sorted([home_team, visitor_team]))

    row["HomeTeamWonLast"] = 1 if last_match_winner[teams] == row['Home/Neutral'] else 0
    dataset.loc[index] = row

    winner = row['Home/Neutral'] if row['HomeWin'] else row['Visitor/Neutral']
    last_match_winner[teams] = winner
```

```python
X_lastwinner = dataset[['HomeTeamRanksHigher', 'HomeTeamWonLast']].values
clf = DecisionTreeClassifier(random_state=14)
scores = cross_val_score(clf, X_lastwinner, y_true, scoring='accuracy')
print(f"The accuracy is {np.mean(scores) * 100} ")
```

```python
from sklearn.preprocessing import LabelEncoder
encoding = LabelEncoder()

encoding.fit(dataset['Home/Neutral'].values)
```

```python
home_teams = encoding.transform(dataset['Home/Neutral'].values)
visitor_teams = encoding.transform(dataset['Visitor/Neutral'].values)
X_teams = np.vstack([home_teams, visitor_teams]).T
```

```python
from sklearn.preprocessing import OneHotEncoder
onehot = OneHotEncoder()
```

```python
X_teams_expanded = onehot.fit_transform(X_teams).todense()
```

```python
clf = DecisionTreeClassifier(random_state=14)
scores = cross_val_score(clf, X_teams_expanded, y_true, scoring='accuracy')
print(f"The accuracy is {np.mean(scores) * 100} ")
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(random_state=14)
scores = cross_val_score(clf, X_teams, y_true, scoring='accuracy')
print(f"The accuracy is {np.mean(scores) * 100} ")
```

```python
X_all = np.hstack([X_homehigher, X_teams])
clf = RandomForestClassifier(random_state=14)
scores = cross_val_score(clf, X_all, y_true, scoring='accuracy')
print(f"The accuracy is {np.mean(scores) * 100} ")
```

```python
from sklearn.model_selection import GridSearchCV
parameter_space = {
    "max_features": [2, 10, 'auto'],
    'n_estimators': [100,],
    'criterion': ['gini', 'entropy'],
    'min_samples_leaf': [2, 4, 6],
}

clf = RandomForestClassifier(random_state=14)
grid = GridSearchCV(clf, parameter_space)
grid.fit(X_all, y_true)
print(f"The accuracy is {grid.best_score_ * 100} ")
```
