```python
import pandas as pd
```

```python
housing = pd.read_csv('./datasets/chapter_2/housing.csv')
```

```python
housing.head()
```

```python
housing['state'].value_counts()
il_housing = housing.query("state == 'il'")
print(il_housing.head())
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
il_housing.hist(bins=50, figsize=(20, 15))
```

```python
il_housing.plot(kind='scatter', x='long', y='lat', alpha=0.4, s='sqfeet', label='square feet', c='price', cmap=plt.get_cmap('jet'), colorbar=True, figsize=(20, 15))
plt.legend()
```

```python
correlation_matrix = il_housing.corr()
```

```python
correlation_matrix['price'].sort_values(ascending=False)
```

```python
from pandas.plotting import scatter_matrix
attributes = ['price', 'sqfeet', 'beds', 'baths', 'type']
scatter_matrix(il_housing[attributes], figsize=(15, 8))
```

```python
il_housing.plot(kind='scatter', x='sqfeet', y='price', alpha=0.1)
```

```python
feature_cols = 'sqfeet'

X = il_housing[[feature_cols]]
y = il_housing['price']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

lin_model = LinearRegression()
lin_model.fit(X_train, y_train)
```

```python
y_train_predict = lin_model.predict(X_train)
```

```python
import numpy as np
rmse = (np.sqrt(mean_squared_error(y_train, y_train_predict)))
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_train, y_train_predict)
print("The model performance for training set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
print("\n")
```

```python
y_test_predict = lin_model.predict(X_test)
rmse = (np.sqrt(mean_squared_error(y_test, y_test_predict)))
r2 = r2_score(y_test, y_test_predict)

print("The model performance for testing set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
```
